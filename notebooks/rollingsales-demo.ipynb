```python
import re
def extract(): 
    with open('./rollingsales_manhattan.csv') as f:
        for line in f:
            line_parts = line.split(',')
            try:
                #price = int(line_parts[-2].replace(',', '').replace('""'))
                price = int(re.sub(r"\"|,", '', line_parts[-2]))
            except ValueError:
                continue
            neighborhood = line_parts[1]
            address = line_parts[8]
            yield neighborhood, address, price
```

```python
list(extract())
```
