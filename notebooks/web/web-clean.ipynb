```python
# can we get some data back?

base = "https://rickandmortyapi.com/api/"
url = f'{base}/character'
def ram_gen(start_url):
    res = urllib.request.urlopen(url)
    s = res.read()
    data = json.loads(s)['results']
    try:
        next_url = result['info']['next']
    except KeyError:
        next_url = None
    next_url = next_url if next_url and 'http' in next_url else None 
    while next_url is not None:
        yield data 
        res = urllib.request.urlopen(next_url)
        s = res.read()
        try:
            data = json.loads(s)['results']
        except KeyError:
            data = None
        try:
            next_url = result['info']['next']
        except KeyError:
            next_url = None
        next_url = next_url if next_url and 'http' in next_url else None 
        print(next_url)
    yield data
    
    
    
    
    
        
```
