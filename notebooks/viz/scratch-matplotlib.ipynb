```python
import random
counter = {}

code_points = list(range(97, 123))
copy_code_points = code_points[:]

base = [chr(n) for n in code_points]
random.shuffle(copy_code_points)
new_alpha =  [chr(n) for n in copy_code_points]

print(base, new_alpha)
with open('files/source.txt', 'r') as fr:
    with open('files/enc.txt', 'w') as fw:
        for line in fr:
            for ch in line:
                if ch.isalpha():
                    norm_ch = ch.lower()
                    new_ch = new_alpha[base.index(norm_ch)]
                    fw.write(new_ch)
                else:
                    fw.write(ch)
```

```python
from collections import Counter
new_counter = Counter()
with open('files/enc.txt', 'r') as f:
    for line in f:
        for ch in line:
            if ch.isalpha():
                norm_ch = ch.lower()
                new_counter[norm_ch] = new_counter.get(norm_ch, 0) + 1
```

```python
letters, counts = zip(*new_counter.most_common())
plt.bar(np.arange(len(letters)), counts)
enc_freq = ''.join(letters)
plt.xticks(np.arange(len(letters)), letters)
```

```python
with open('files/enc.txt', 'r') as fr:
    with open('files/solved.txt', 'w') as fw:
        for ch in fr.read():
            if ch.isalpha():
                ch = freq[enc_freq.index(ch)]
            fw.write(ch)
```

```python
import urllib
```

```python
response = urllib.request.urlopen('https://openei.org/doe-opendata/dataset/a7fea769-691d-4536-8ed3-471e993a2445/resource/86c50aa8-e40f-4859-b52e-29bb10166456/download/populationbycountry19802010millions.csv').read().decode('utf-8')
```

```python
rows_temp = response.split('\r')
```

```python
#print(rows_temp)
table = [row.split(',') for row in rows_temp]
def convert(row):
    result = []
    for ele in row:
        try:
            n = float(ele)
        except ValueError:
            n = np.nan
        result.append(n)
    return result
        
table_dict = {row[0]: convert(row[1:]) for row in table[1:]}
#print(table_dict)
years = [int(y) for y in table[0] if y.isnumeric()]
# print(len(years), years)
for country, pops in table_dict.items():
    delta = (pops[-1] - pops[0]) / pops[-1]
    #print(country, delta)
    if delta > 0.7:
        try:
            plt.plot(years, pops, label=country)
        except ValueError:
            print('value error')
            pass
            #print(country, len(pops), pops)
plt.legend()
```
