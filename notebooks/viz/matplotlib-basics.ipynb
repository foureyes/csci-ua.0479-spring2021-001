### Simple Line Plots using `plot`

```python
x = np.arange(10)
y = x ** 2
plt.plot(x, y)
```

```python
plt.plot(x, y, 'r^')
plt.plot(x, x, 'k')
plt.plot(x, x ** 3, 'g.')
```

```python
plt.plot(x, y, x, x, x, x + 5)
plt.xlabel('hours of sleep')
plt.ylabel('number of good decisions made')
plt.xticks([0, 4, 9], ['no sleep!', 'some sleep', '2 much'])
plt.yticks([0, 80], ['all mistakes', 'flawless decisions'])
```
