```python
sizes = [5, 7, 1, 3]
labels = ['strawberry', 'apple', 'cherry', 'humble']
colors=['red', (0, 0.75, 0.5), '#aaaaaa', 'b']
plt.axis('equal')
plt.pie(sizes, labels=labels, autopct='%.2f%%', colors=colors)
```

```python
feels = {'thumbs up': 12, 'suspicious': 6, 'potato':8}
sizes = feels.values()
labels = feels.keys()
plt.axis('equal')

def make_autopct(vals):
    return lambda pct: f'{pct:.2f}% ({int(0.01 * pct * sum(vals))})'

plt.pie(sizes, labels=labels, autopct=make_autopct(sizes))
```

```python
from collections import Counter
counts = {}
with open('files/enc.txt') as f:
    for line in f:
        for ch in line:
            if ch.isalpha():
                counts[ch.lower()] = counts.get(ch.lower(), 0) + 1
                
print(sorted(counts.items(), key=lambda t: t[1], reverse=True)[0])
print(sorted(counts.items(), key=lambda t: t[1], reverse=True))

from collections import Counter
counts = Counter()
with open('files/enc.txt') as f:
    for line in f:
        for ch in line:
            if ch.isalpha():
                counts[ch.lower()] += 1
print(counts)
print(counts.most_common(1))
with open('files/enc.txt') as f:
    counts = Counter(ch.lower() for ch in f.read() if ch.isalpha())
print(counts)


plt.subplot(2, 1, 1)
plt.bar(np.arange(len(counts)), counts.values(), )
plt.xticks(np.arange(len(counts)), counts.keys())
#plabels=counts.keys()

plt.subplot(2, 1, 2)
ordered = counts.most_common()
plt.bar(np.arange(len(ordered)), [t[1] for t in ordered])
plt.xticks(np.arange(len(ordered)), [t[0] for t in ordered])
#plabels=counts.keys()


freq_source = ''.join(list(zip(*counts.most_common()))[0])
freq_target = 'etaonisrhdlumcfwgpybvkxjqz' 
print(freq_source)
print(freq_target)
with open('files/enc.txt') as f:
    for line in f:
        s = ''
        for ch in line:
            ch = ch.lower()
            if ch.isalpha():
                i = freq_source.index(ch)
                s += freq_target[i]
            else:
                s += ch
        print(s)
```
